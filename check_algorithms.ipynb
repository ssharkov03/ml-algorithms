### Linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from Linear.linear_regression import LinearRegression

sns.set_theme()

l = 100
X = np.linspace(0, 1000, l).reshape((l, 1))
y = np.array([3 * x - 50 + np.random.normal(loc=0, scale=100) for x in X])

target = lambda x: 3 * x - 50

linreg = LinearRegression(batch_size=10,
                          learning_rate=1e-6,
                          num_iter=100,
                          verbose=2)
linreg.fit(X, y)
pred = linreg.predict(X)
weights = linreg.get_weights()

plt.scatter(X, y)
plt.plot(X, target(X.ravel()), label="Target: y = 3x - 50", color="r")
plt.plot(X, pred, label=f"Predicted: y = {weights['w']:0.1f}x + {weights['bias']:0.1f}", color="green")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()

```
